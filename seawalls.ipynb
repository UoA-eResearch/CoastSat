```python
import geopandas as gpd
from tqdm.auto import tqdm
import pandas as pd
seawalls = gpd.read_file("aklc_seawalls.geojson")
seawalls
```

```python
transects = gpd.read_file("transects_extended.geojson")
transects
```

```python
intersecting_transects = transects[transects.intersects(seawalls.union_all())]
intersecting_transects
```

```python
intersection_points = intersecting_transects.intersection(seawalls.union_all())
intersection_points
```

```python
intersection_points.geom_type.value_counts()
```

```python
def reduce_to_furthest_point(idx, geom):
    if geom.geom_type == "MultiPoint":
        line = intersecting_transects.loc[idx, "geometry"]
        return max(geom.geoms, key=line.project)
    return geom

intersection_points = gpd.GeoSeries(
    [reduce_to_furthest_point(idx, geom) for idx, geom in intersection_points.items()],
    index=intersection_points.index,
    crs=intersection_points.crs,
    name=intersection_points.name,
)

intersection_points.geom_type.value_counts()
```

```python
intersecting_transects["origin_distance_to_seawall"] = intersecting_transects.to_crs(2193).project(intersection_points.to_crs(2193))
intersecting_transects["origin_distance_to_seawall"].describe()
```

```python
intersecting_transects.set_index("id").origin_distance_to_seawall
```

```python
results = []
for site_id in tqdm(intersecting_transects.site_id.unique()):
    df = pd.read_csv(f"data/{site_id}/transect_time_series_tidally_corrected.csv").set_index("dates")
    results.append((df - intersecting_transects.set_index("id").origin_distance_to_seawall).dropna(axis=1, how="all"))
results = pd.concat(results, axis=1).sort_index()
results.to_csv("seawall_distance.csv")
results
```
